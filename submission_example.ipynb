```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
```

```python
columns_to_include = [
    "MS SubClass",
    "MS Zoning",
    "Lot Area",
    "Utilities",
    "Lot Config",
    "Neighborhood",
    "Condition 1",
    "Condition 2",
    "Bldg Type",
    "House Style",
    "Overall Qual",
    "Overall Cond",
    "Year Built",
    "Year Remod/Add",
    "Exterior 1st",
    "Exterior 2nd",
    "Exter Qual",
    "Exter Cond",
    "Foundation",
    "Heating",
    "Heating QC",
    "Central Air",
    "Electrical",
    "1st Flr SF",
    "2nd Flr SF",
    "Gr Liv Area",
    "Full Bath",
    "Half Bath",
    "Bedroom AbvGr",
    "Kitchen AbvGr",
    "Kitchen Qual",
    "Functional",
    "Garage Type",
    "Garage Cars",
    "Wood Deck SF",
    "Open Porch SF",
    "Pool Area",
    "Misc Val",
    "Mo Sold",
    "Yr Sold",
    "Sale Type"
]

categorical = ["MS SubClass",
               "MS Zoning",
               "Lot Config",
               "Neighborhood",
               "Condition 1",
               "Condition 2",
               "Bldg Type",
               "House Style",
               "Exterior 1st",
               "Exterior 2nd",
               "Foundation",
               "Heating",
               "Central Air",
               "Electrical",
               "Garage Type",
               "Mo Sold",
               "Yr Sold",
               "Sale Type"]
categorical_2 = []
to_convert_numbers = ["Utilities",
                      "Exter Qual",
                      "Exter Cond",
                      "Heating QC",
                      "Kitchen Qual",
                      "Functional"
]

rating_conversion = {
    "AllPub": 3,
    "NoSewr": 2,
    "NoSeWa": 1,
    "ELO": 0,
    "Ex": 4,
    "Gd": 3,
    "TA": 2,
    "Fa": 1,
    "Po": 0,
    "Typ": 7,
    "Min1": 6,
    "Min2": 5,
    "Mod": 4,
    "Maj1": 3,
    "Maj2": 2,
    "Sev": 1,
    "Sal": 0
}
```

```python
train = pd.read_csv('datasets/train.csv')
test = pd.read_csv('datasets/test.csv')
```

```python
len(train.columns)
```

```python
len(test.columns)
```

```python
filtered_train = train[columns_to_include]
filtered_test = test[columns_to_include]
```

```python
len(filtered_train.columns)
```

```python
len(filtered_test.columns)
```

```python
categorical_train = filtered_train[categorical]
ohc = OneHotEncoder(handle_unknown='ignore', drop='first')
ohc.fit(categorical_train)
encoded_train = pd.DataFrame(ohc.transform(categorical_train).toarray(), columns=ohc.get_feature_names_out())
```

```python
filtered_train.drop(columns=categorical, inplace=True)
filtered_train = pd.concat([filtered_train, encoded_train], axis=1)
filtered_train.head()
```

```python
categorical_test = filtered_test[categorical]
encoded_test = pd.DataFrame(ohc.transform(categorical_test).toarray(), columns=ohc.get_feature_names_out())
filtered_test.drop(columns=categorical, inplace=True)
filtered_test = pd.concat([filtered_test, encoded_test], axis=1)
filtered_test.head()
```
